### Load Packages and Set Environment

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import seaborn as sns; sns.set()

%matplotlib inline
```

### User Defined Functions

```python
# Creates samples for a data profiler by getting up to 5 unique values
def create_sample(data_set, column):
    data_copy = data_set[[column]].copy()
    data_copy['sample'] = data_copy[column]
    data_copy['sample'] = data_copy['sample'].astype(str)
    data_copy['sample'] = data_copy['sample'].apply(lambda x: x[:150])
    sample_list = data_copy['sample'].unique()
    sample_string = ", ".join(sample_list[:min(len(sample_list),5)])
    return sample_string

# Only should be used for temperature and preciptation since they the times are in order.
# Locates null values and takes average between preceeding and following value
def null_filler(df, column):
    null_indexes = df[column][df[column].isnull()].index
    
    for i in null_indexes:
        if i == (df.shape[1] - 1):
            df[column][i] = df[column][i-1]
        elif i == 0:
            df[column][i] = df[column][i+1]
        else:
            df[column][i] = (df[column][i-1] + df[column][i+1])/2
            
    return df[column]
```

### Loading Dataset, Initial Analysis

```python
data = pd.read_csv('Brooklyn_Bridge_Automated_Pedestrian_Counts_Demonstration_Project.csv')
```

### Create a data profiler

```python
data_profiler = pd.DataFrame(data = data.columns, columns = ['columns'])
data_profiler['dtype'] = data.dtypes.values
data_profiler['unique_values'] = data.nunique().values
data_profiler['nulls'] = data.isnull().sum().values
data_profiler['sample'] = data_profiler['columns'].apply(lambda x: create_sample(data_set=data, column=x))
```

```python
data_profiler
```

### Subsetting and adjusting columns
To help reduce some of the columns to look at, we'll only look at columns with more than unique value

```python
unique_values = data_profiler.loc[data_profiler['unique_values'] > 1,'columns'].values
data_trim = data.loc[:, unique_values]
```

```python
# Filling in to be used in groupby's later on
data_trim['events'] = data_trim['events'].fillna('')
```

```python
# Filling in null values
data_trim['temperature'] = null_filler(data_trim, 'temperature')
data_trim['precipitation'] = null_filler(data_trim, 'precipitation')
```

```python
# Adds buckets based on Pedestrian Count to help with visualizations later on
data_trim['Pedestrians_Bin'] = pd.cut(x = data_trim['Pedestrians'], bins = [-1,10,100,1000,10000])
```

### Date Extractions

```python
# Convert hour_beginning to date time object. Easier to work with.
data_trim['hour_beginning_dt'] = pd.to_datetime(data_trim['hour_beginning'])
data_trim['date'] = data_trim['hour_beginning_dt'].apply(lambda x: x.date())
data_trim['year_month'] = data_trim['hour_beginning_dt'].apply(lambda x: str(x.year).zfill(4) + '-' + str(x.month).zfill(2))
data_trim['year'] = data_trim['hour_beginning_dt'].apply(lambda x: x.year)
data_trim['month'] = data_trim['hour_beginning_dt'].apply(lambda x: x.month)
data_trim['day'] = data_trim['hour_beginning_dt'].apply(lambda x: x.day)
data_trim['hour'] = data_trim['hour_beginning_dt'].apply(lambda x: x.hour)
data_trim['day_of_week'] = data_trim['hour_beginning_dt'].apply(lambda x: x.weekday())
```

### Plotting Traffic Over Time

```python
daily_traffic = data_trim.groupby(['date', 'events']).agg({'Pedestrians': 'sum'}).reset_index()

# Adding an exponentially weighted line to help visualize through noise
daily_traffic['Smoothed'] = daily_traffic['Pedestrians'].ewm(alpha = 1/7.).mean()
```

```python
figure, ax = plt.subplots(ncols = 1, figsize = (12,6))
g = sns.scatterplot(x = 'date', y = 'Pedestrians', data = daily_traffic, alpha = 0.8, label = 'Daily Count')
g = sns.lineplot(x = 'date', y = 'Smoothed', data = daily_traffic, label = 'Smoothed Count', linestyle = "-", color = 'orange')
g = sns.scatterplot(x = 'date', y = 'Pedestrians', data = daily_traffic.loc[daily_traffic['events']!= ''], label = 'Events', color = 'red', marker = 'x')
plt.xlim(datetime.date(2017, 10, 1), datetime.date(2018, 7, 31))
plt.ylabel('Pedestrian Count')
plt.xlabel('Date')
plt.title('Daily Sums of Pedestrian Traffic')
plt.xticks(rotation = 90)
plt.show()
```

```python
monthly_traffic = data_trim.groupby(['year_month']).agg({'Pedestrians': 'mean'}).reset_index()
```

```python
figure, ax = plt.subplots(ncols = 1, figsize = (12,6))
g = sns.lineplot(x = 'year_month', y = 'Pedestrians', data = monthly_traffic)
plt.xticks(rotation = 90)
plt.title('Monthly Averages of Pedestrians')
plt.show()
```
